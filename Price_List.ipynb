```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from scipy import stats
import re
from tqdm import tqdm
```

```python
def get_model_price(url_):
    url = url_
    at = re.compile('>.*<')
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'lxml')

    cl = 'productSKULink ellips-line nav--snhdr ttl-height m-b-xs txt-al-c m-t-m txt-m ellips-line-ell-2 product_link-60494'
    cl1 = 'col-12 v-al-mdl p-h-xs'
    pri = 'd-block txt-al-c'
    vfm_cl = 'vfm__scr'
    mob = soup.findAll('div',{'class':cl1})
    price = soup.findAll('span',{'class':pri})
    vfm = soup.findAll('span',{'class':vfm_cl})
    for r in mob:
        model_names.append(r.find('a').getText())
    
    for p in price:
        price_val.append(p.getText())
        
    for v in vfm:
        vfm_list.append(v.getText())
    
    return price_val, model_names,vfm_list


def get_pages(url):
    if(url not in visited):
        
        page_di[1] = url
        count = 0
        r = requests.get(url)
        soup = BeautifulSoup(r.content, 'lxml')
        pages = soup.findAll('a',{'class':'pgntn-lnk'},href=True)
        print('Url Visiting :',url)
        visited.append(url)
        for p in tqdm(pages):

            st = 'https://pricebaba.com'
            pg_link = st+str(p['href'])
            pg_index = re.findall('page=([0-9]+)',p['href'])
            print(pg_index)
            if(len(pg_index)==0):
                if(p.getText().isdigit()):
                    pg_index = p.getText()
                    print(pg_index)
                    ind = int(pg_index)
                    page_di[ind] =  pg_link
                    if(pg_link not in visited):
                        page_di[ind] = pg_link 
                        get_pages(pg_link)
            else:
                ind = int(pg_index[0])
                page_di[ind] = pg_link
                if(pg_link not in visited):
                    page_di[ind] = pg_link 
                    get_pages(pg_link)
          


r1 = input('Enter range 1:')
r2 = input('Enter range 2 :')
page_di = {}
visited=[]
model_names = []
price_val = []
vfm_list = []
pg = []
url = 'https://pricebaba.com/search?limit=40&online_price='+str(r1)+'-'+str(r2)+'&status=20&category=MOBILE&status=20&start=0&default=true&sort=vfm_score-desc'
get_pages(url)
for url in tqdm(page_di.values()):
    get_model_price(url)
    print('model',len(model_names))
    print('price',len(price_val))
    print('vfm',len(vfm_list))
    model = model_names
    price = price_val
    vfm = vfm_list
model_price=list(zip(model,price,vfm))
df=pd.DataFrame(model_price)
df.columns = ['Device_Name','Device_Price','VFM']    


```
